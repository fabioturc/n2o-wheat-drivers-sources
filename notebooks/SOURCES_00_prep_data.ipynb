# LOAD LIBRARIES

```python
import pandas as pd
import numpy as np
```

# LOAD DATA

```python
n2o_data  = pd.read_csv('../data/raw/n2o_isotopes.csv')
soil_data = pd.read_csv('../data/raw/soil_water_isotopes.csv')
meteo     = pd.read_csv('../data/raw/meteo.csv')

display(n2o_data)
display(soil_data.head())
display(meteo.head())
```

# PROCESSING

Extract START/END from sample_ID_IRMS and parse date (dd/mm/yyyy)

```python
n2o_data['start_end'] = n2o_data['sample_ID_IRMS'].astype(str).str.extract(r'(START|END)')[0]
n2o_data['date'] = pd.to_datetime(n2o_data['date'], format='%d/%m/%Y', errors='coerce').dt.normalize()

display(n2o_data[['sample_ID_IRMS', 'start_end', 'date']])
```

Reshape

```python
value_cols = [
    'CO2.TCD.ppm', 'N2O.ECD.ppm', 'CH4.FID.ppm',
    'd15Nbulk', 'd15Nalpha', 'd15Nbeta',
    'SP', 'd18O'
]

id_cols = [c for c in n2o_data.columns if c not in (['sample_ID_IRMS'] + value_cols + ['start_end'])]

wide = (
    n2o_data.drop(columns=['sample_ID_IRMS'])
    .pivot_table(index=id_cols, columns='start_end', values=value_cols, aggfunc='first')
)

# Flatten MultiIndex columns: (variable, START/END) -> variable_START / variable_END
wide.columns = [f"{var}_{se}" for var, se in wide.columns]
n2o_wide = wide.reset_index()

display(n2o_wide)
print('Columns:', list(n2o_wide.columns))
```

**Keeling**

Computation of Keeling-like SP, Î´Â¹âµNbulk, Î´Â¹â¸O of the emitted N2O.

```python
n2o_data = n2o_wide.copy()

# Î”N2O in ppb (ppm difference Ã— 1000)
n2o_data['diff_n2o_ppb'] = (n2o_data['N2O.ECD.ppm_END'] - n2o_data['N2O.ECD.ppm_START']) * 1000

# Keeling-style mixing equations
den = (n2o_data['N2O.ECD.ppm_END'] - n2o_data['N2O.ECD.ppm_START'])
n2o_data['SP']   = (n2o_data['SP_END']       * n2o_data['N2O.ECD.ppm_END'] - n2o_data['SP_START']       * n2o_data['N2O.ECD.ppm_START']) / den
n2o_data['d15N'] = (n2o_data['d15Nbulk_END'] * n2o_data['N2O.ECD.ppm_END'] - n2o_data['d15Nbulk_START'] * n2o_data['N2O.ECD.ppm_START']) / den
n2o_data['d18O'] = (n2o_data['d18O_END']     * n2o_data['N2O.ECD.ppm_END'] - n2o_data['d18O_START']     * n2o_data['N2O.ECD.ppm_START']) / den

display(n2o_data[['date', 'chamber', 'diff_n2o_ppb', 'SP', 'd15N', 'd18O']].head())
```

**Flux calcs**

- Take meteo at 13:15 (TA, PA)
- Ideal gas conversion using PA (must be **Pa**) and TA (Â°C)
- Convert Î”N2O(ppb) â†’ mol m-3, then to surface flux using chamber volume/area and closure time (1 h).
- Calculate flux in nmol m-2 s-1

```python
# Parse TIMESTAMP_MIDDLE and filter the station at 13:15
meteo['TIMESTAMP_MIDDLE'] = pd.to_datetime(meteo['TIMESTAMP_MIDDLE'], errors='coerce').dt.tz_localize(None)

meteo_1315 = (
    meteo.loc[meteo['TIMESTAMP_MIDDLE'].dt.strftime('%H:%M') == '13:15', ['TIMESTAMP_MIDDLE', 'TA', 'PA']]
    .assign(date=lambda d: d['TIMESTAMP_MIDDLE'].dt.normalize())
    .drop(columns=['TIMESTAMP_MIDDLE'])
)

# Join meteo to N2O table by date
n2o_data = n2o_data.merge(meteo_1315, on='date', how='left')

# Constants
R = 8.314        # J mol^-1 K^-1
t = 3600         # s (1 hour)

# Chamber geometry
h_chamber = 0.25 # m
r_chamber = 0.15 # m
A = np.pi * r_chamber**2   # m^2
V = A * h_chamber          # m^3

# IMPORTANT: PA must be in Pa

n2o_data['delta_n2o_mol_per_m3'] = (n2o_data['diff_n2o_ppb'] * n2o_data['PA']) / (R * (273.15 + n2o_data['TA']) * 1e9)
n2o_data['flux_nmol'] = (n2o_data['delta_n2o_mol_per_m3'] * V) / (A * t) * 1e9
```

Keep selected columns

```python
n2o_keep = n2o_data[['date', 'chamber', 'SP', 'd15N', 'd18O', 'diff_n2o_ppb', 'flux_nmol']].copy()
n2o_keep
```

Get soil water isotopes

```python
soil = soil_data.copy()
soil['date'] = pd.to_datetime(soil['date'], format='%d.%m.%Y', errors='coerce').dt.normalize()
soil['d18O_water'] = soil['d18O_avg']
soil_keep = soil[['date', 'chamber', 'd18O_water']].copy()
soil_keep
```

Merge water isotope data with N2O isotope data

```python
iso_data = n2o_keep.merge(soil_keep, on=['date', 'chamber'], how='left')
iso_data
```
